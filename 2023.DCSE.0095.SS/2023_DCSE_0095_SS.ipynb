```python
from empiricaldist import Pmf
```

```python
Weight = df['Weight']
```

```python
pmfWeight = Pmf.from_seq( Weight, normalize=False)
type(pmfWeight)
```

```python

pmfWeight.head()
```

```python
pmfWeight.tail()
```

```python
pmfWeight[70]
```

```python

pmfWeight_norm = Pmf.from_seq(Weight, normalize=True)
pmfWeight_norm.head()
```

```python
pmfWeight_norm[70]
```

```python
# pmf provides a bar method that plots the values and their probabilities as a bar chart.
pmfWeight_norm.bar(label='student Weight')
plt.xlabel('Weight in kgs')
plt.ylabel('PMF')
plt.title('A  bar graph showing distribution of weight among cohort 3 students')
plt.legend()
```
