```python
height = df['height']
height = height.astype('str')
print(type(height))
```

```python
weight = df['weight']
weight = weight.astype('str')
print(type(weight))
```

```python

plt.plot(height, weight, "o", alpha=0.5, linestyle="-", color="green"  )
plt.xlabel("height (cm)")
plt.ylabel("weight (Kilograms)")
plt.title("RELATIONSHIP BETWEEN HEIGHT AND WEIGHT OF COHORT THREE")
plt.xlim(6)
plt.ylim(6)
```

```python
plt.xlabel("weight")
plt.ylabel("height")
plt.plot(height = df['height'],weight = df['weight'],)
plt.title("RELATIONSHIP BETWEEN HEIGHT AND WEIGHT OF COHORT THREE")
```

```python
#title
plt.xlabel("weight in kilograms")
plt.ylabel("height in centimeters")
plt.plot(height = df['height'],weight = df['weight'],)
plt.title("A scatter plot showing relationship between weight and height of cohort 3 students of 2023")
```
